# Inconsistent Spelling Check

```r
library(tidyverse)
```

```r
filePath <- "./data/inspections.csv"
```

```r
names <- c('InspectionID', 'RestaurantName', 'OtherName', 'LicenseNumber', 'FacilityType', 'Risk', 
           'Address', 'City', 'State', 'ZIP', 'InspectionDate', 'InspectionType', 'Results',
           'Violations', 'Latitude', 'Longitude', 'Location')
```

```r
inspections <- read_csv(filePath, col_names=names, skip=1)
```

```r
inspections %>%
    group_by(RestaurantName) %>%
    summarize(Frequency=n()) %>%
    arrange(desc(Frequency)) %>%
    head(7)
```

```r
inspections %>%
    filter(grepl('McD', RestaurantName, ignore.case=TRUE))%>%
    select(RestaurantName) %>%
    unique()
```

```r
junk_words <- inspections %>%
    filter(grepl('McD', RestaurantName, ignore.case=TRUE))%>%
    filter(RestaurantName != 'SARAH MCDONALD STEELE') %>%
    select(RestaurantName) %>%
    unique() %>%
    pull(RestaurantName) # Pull out RestaurantName as a vector
```

```r
junk_words
```

```r
inspections %>%
    mutate(RestaurantName=ifelse(RestaurantName %in% junk_words, 'MCDONALD', RestaurantName)) %>%
    group_by(RestaurantName) %>%
    summarize(Freq=n()) %>%
    arrange(desc(Freq))
```
